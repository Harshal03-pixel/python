```python
def swap_with_temp(a, b):
    print(f"Before swapping: a = {a}, b = {b}")

    # Using a temporary variable to swap values
    temp = a
    a = b
    b = temp

    print(f"After swapping: a = {a}, b = {b}")

swap_with_temp(5, 10)

```

17) What are negative indexes and why are they used?

Negative indexes in Python are a way to access elements from the end of a sequence, such as a list, string, or tuple. Python allows negative indexing to provide a more convenient way to work with elements at the end of these sequences without needing to know their length in advance.

**How Negative Indexing Works:**

**Positive Indexing**: Starts from the beginning of the sequence. The first element is at index **0**, the second element is at index **1** and so on.

**Negative Indexing**: Starts from the end of the sequence. The last element is at index **-1**, the second-to-last element is at index **-2**, and so on.

**Example:**


```python
my_list = ['a', 'b', 'c', 'd', 'e']

# Positive indexing
print(my_list[0])  # Output: 'a'
print(my_list[1])  # Output: 'b'

# Negative indexing
print(my_list[-1])  # Output: 'e'
print(my_list[-2])  # Output: 'd'
```
